## `.loc` dan `.iloc`

Dengan menggunakan `.loc` dan `iloc` kita dapat melakukan pengirisan pada index **baris dan kolom**. 

Perbedaan yang mendasar dari kedua operator ini adalah:
- `.iloc` merujuk pada lokasi **index** baris atau kolomnya sehingga harus **integer**, sedangkan
- `.loc` merujuk pada lokasi **nama** baris atau kolomnya

Syntax: `df.iloc[baris, kolom]` atau `df.loc[baris, kolom]`

## Conditional Subsetting

Selain menggunakan `.loc` dan `.iloc`, kita dapat melakukan subsetting berdasarkan kondisi tertentu. Misal pada dataframe `df_household`, kita ingin mengambil beberapa data dengan kondisi sebagai berikut:

- Transaksi yang terjadi di supermarket: `.format == 'supermarket'`
- Transaksi dengan produk yang harganya lebih dari sama dengan 200000: `.unit_price >= 200000`
- Transaksi dimana kuantitas tidak sama dengan 0: `.quantity != 0`

Syntax penulisan untuk conditional subsetting adalah:

**`df[df['column_name'] <comparison_operator> <value>]`**

atau

**`df[df.column_name <comparison_operator> <value>]`**

Contoh comparison_operator adalah seperti `==`, `!=`, `>`, `>=`, `<`, `<=`.

````python
df_household.head()
````

````python
df_household[['sub_category', 'category']]
````

````python
df_household.category
````

````python
df_household.category.unique
````

````python
df_household[df_household['category'] == 'Rice']
````

````python
df_household[df_household['format'] == 'supermarket']
````

### Multiple Condition

````python
df_household[(df_household['format'] == 'supermarket') | (df_household['unit_price'] >= 50000)]
````

Ambil yang format nya hanya `supermarket`

````python
# df_household[]
# df_household[df_household['format']]
# df_household[df_household['format'] == 'supermarket']

df_household[df_household['format'] == 'supermarket']
````

````python
df_household[(df_household['format'] == 'supermarket') & (df_household['unit_price'] >= 50000) & (df_household['unit_price'] <= 100000)]
````

````python
#Filter data berdasarkan kategori Rice dan format Supermarket
df_household[(df_household['category'] == 'Rice') & (df_household['format'] == 'supermarket')]
````

````python
#Filter data berdasarkan kategori Rice dan format Sugar/Flavored Syrup
df_household[(df_household['category'] == 'Rice') | (df_household['category'] == 'Sugar/Flavored Syrup')]
````

````python
#Cara ke 2 menggunakan .isin
df_household[df_household['category'].isin(['Rice', 'Sugar/Flavored Syrup'])]
````

````python
#Filter Unit Price antara 50.000 dan 100.000
df_household[(df_household['unit_price'] > 50000) & (df_household['unit_price'] < 100000)]
````

````python
df_household[df_household['format'].isin(['supermarket', 'minimarket'])]
#namatable[kondisinyagimana]
````

Filter `unit_price` yang lebih besar sama dengan 200000

````python
df_household[df_household['unit_price'] >= 200000]
````

Kita juga dapat menggunakan operator `&` (AND) dan `|` (OR) untuk melakukan subsetting lebih dari 1 kondisi. Misalnya kita ingin melihat data penjualan dari seorang pegawai bernama Moana yang jumlahnya lebih dari 5000, maka kita dapat menggunakan syntax:
```
sales[(sales.salesperson == 'Moana') & (sales.amount > 5000)]
```

Untuk subsetting dengan kondisi lebih dari 1, setiap kondisi diletakkan **di dalam tanda kurung `()`** atau bisa ditulis dengan syntax berikut:

```
df[(kondisi pertama) operator (kondisi kedua) operator (kondisi ketiga) dan seterusnya...]
```

Poin:
- Operator AND: harus semua kondisi terpenuhi dalam satu baris agar muncul
- Operator OR: salah satu kondisi saja sudah terpenuhi maka baris tsb muncul
- Apabila lebih dari 1 kondisi, maka bungkus kondisi dengan `()`

Misal kita mau melakukan filtering untuk mengambil semua transaksi di `supermarket` yang `unit_price`nya di atas 200000.

````python
df_household[(df_household['format'] == 'supermarket') & (df_household['unit_price'] > 200000)]
````

Filter baris yang `format`nya `supermarket` dan `minimarket`

````python
df_household[(df_household['format'] == 'supermarket') | (df_household['format'] == 'minimarket')]
````

Alternatif menggunakan method `.isin()`

````python
df_household[df_household['format'].isin(['supermarket', 'minimarket'])]
````
